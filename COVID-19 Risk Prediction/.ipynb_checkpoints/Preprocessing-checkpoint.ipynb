```python
import pandas as pd
import numpy as np
import os
import scipy.io
```

# Importing dataset

```python
path = os.path.join( os.getcwd(), '../..', 'data/preprocessed_BMED8813/' )
filename = path + 'TestFeatures.csv'
TestFeatures = pd.read_csv(filename)
filename = path + 'TestLabels.csv'
TestLabels = pd.read_csv(filename)
filename = path + 'TrainFeatures.csv'
TrainFeatures = pd.read_csv(filename)
filename = path + 'TrainLabels.csv'
TrainLabels = pd.read_csv(filename)
```

```python
pd.set_option("display.max_rows", 10)
TrainLabels.columns
```

```python
TrainSet = TrainFeatures.join(TrainLabels)
TestSet = TestFeatures.join(TestLabels)
```

# Selecting features for imputation

```python
features_toImpute = ['averageBMI','glucoseMedian','glucoseLast','SodiumMedian','SodiumLast','ALTMedian','ALTLast',
                    'plateletsMedian','plateletsLast','GFRMedian','GFRLast','CreatinineMedian','CreatinineLast']
print(f'Training set has {TrainSet.shape[0]} data points before removing outliers.')
print(f'Testing set has {TestSet.shape[0]} data points before removing outliers.')
```

## Removing outliers from dataset

```python
for feature in features_toImpute:
    lower_bound = TrainSet[feature].quantile(.003)
    upper_bound = TrainSet[feature].quantile(.997)
    TrainSet.drop(TrainSet[(TrainSet[feature] <= lower_bound)|(TrainSet[feature] >= upper_bound)].index, inplace=True)
    TrainSet = TrainSet.reset_index(drop=True)
    
for feature in features_toImpute:
    lower_bound = TestSet[feature].quantile(.003)
    upper_bound = TestSet[feature].quantile(.997)
    TestSet.drop(TestSet[(TestSet[feature] <= lower_bound)|(TestSet[feature] >= upper_bound)].index,inplace=True)
    TestSet = TestSet.reset_index(drop=True)
```

```python
print(f'Training set has {TrainSet.shape[0]} data points after removing outliers.')
print(f'Testing set has {TestSet.shape[0]} data points after removing outliers.')
```

## Imputation of continuous variables with KNN

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=5)
imputed_data = imputer.fit_transform(TrainSet)  # impute all the missing data
df_temp = pd.DataFrame(imputed_data)
df_temp.columns = TrainSet.columns
```

```python
TrainSet[features_toImpute] = df_temp[features_toImpute]
```

```python
imputed_data = imputer.fit_transform(TestSet)  # impute all the missing data
df_temp = pd.DataFrame(imputed_data)
df_temp.columns = TestSet.columns
```

```python
TestSet[features_toImpute] = df_temp[features_toImpute]
```

```python
#Checking missing values after imputation of continuous features
display('Training set missing features:',TrainSet.isna().sum()[TrainSet.isna().sum()!=0])
display('Test set missing features:',TestSet.isna().sum()[TestSet.isna().sum()!=0])
```

## Imputation of remaining values (categorical variables)

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imputed_data = imputer.fit_transform(TrainSet)
df_temp = pd.DataFrame(imputed_data)
df_temp.columns = TrainSet.columns
TrainSet = df_temp
```

```python
#Checking there are no missing values on training set
if TrainSet.isna().sum()[df_temp.isna().sum()!=0].empty:
    display('No missing values')
```

```python
TrainSet.isnull().values.any()
```

```python
#Checking for outliers
import seaborn as sns
ax1 = sns.boxplot(data=TrainSet[features_toImpute], orient="h", palette="Set2")
```

```python
ax2 = sns.boxplot(data=TestSet[features_toImpute], orient="h", palette="Set2")
```

## Applying z-score for dataset standardization

```python
features_toScale = ['averageBMI','glucoseMedian','glucoseLast','SodiumMedian','SodiumLast','ALTMedian','ALTLast',
                    'plateletsMedian','plateletsLast','GFRMedian','GFRLast','CreatinineMedian','CreatinineLast']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
TrainFeatures_scaled = pd.DataFrame(scaler.fit_transform(TrainSet[features_toScale]), columns = TrainSet[features_toScale].columns)
TrainFeatures_scaled
```

```python
TrainSet_final = TrainSet.copy().reset_index(drop=True)
TrainSet_final.loc[:,features_toScale] = TrainFeatures_scaled
TrainSet_final
```

```python
TestFeatures_scaled = pd.DataFrame(scaler.fit_transform(TestSet[features_toScale]), columns = TestSet[features_toScale].columns)
TestFeatures_scaled.head()
```

```python
TestSet_final = TestSet.copy().reset_index(drop=True)
TestSet_final.loc[:,features_toScale] = TestFeatures_scaled
TestSet_final
```

```python
ax1 = sns.boxplot(data=TrainSet_final[features_toImpute], orient="h", palette="Set2")
```

## Saving processed dataset

```python
path = os.path.join( os.getcwd(), '../..', 'data/preprocessed_BMED8813/processed data/' )
filename = path+'TrainSet0501.csv'
TrainSet_final.to_csv(r''+filename,index=False)
filename = path+'TestSet0501.csv'
TestSet_final.to_csv(r''+filename,index=False)
```
